```python
#drop column 'Country ISO3' and 'Indicator Code'
df.drop(['Country ISO3', 'Indicator Code'], axis=1, inplace=True)
df.head()
```

```python
#add new column indicator code and for each unique indicator name, add a unique indicator code as IND_01
df['Indicator Code'] = df['Indicator Name'].apply(lambda x: 'IND_' + str(df['Indicator Name'].unique().tolist().index(x) + 1))
df
```
